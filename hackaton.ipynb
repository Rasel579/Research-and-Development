1)	Endpoint сервисов задается в виде [base url]/сервис, где [base url] это [http://][адрес:порт][/путь]
2)	Все методы POST
3)	Необходимо реализовать 10 REST API сервисов:
	/object-exists
	/get-object-name
	/is-connected
	/find-connected
	/find-common-connected
	/find-exclusive-connected
	/count-all-connected
	/count-connected
	/find-by-sound-match
	/find-connected-by-sound-match

4)	Все параметры в запросах (см. таблицу 1) обязательные
5)	Все ответы всегда только с кодом 200 в случае корректного запроса или 400 в случае некорректных параметров запроса (некорректные ID, типа, параметров).
6)	Для 400 возвращается ответ с пустым телом.
7)	Описание методов приведено в таблице 1.
