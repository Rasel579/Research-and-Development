```python
import json

file_path = './raw_data/text_neighborhood_chat_tg_data.json'
with open(file_path, 'r', encoding='utf-8') as json_file:
    data = json.load(json_file)
data['messages']
```

```python
input_messages = []
for message in data['messages']:
    if message['type'] != 'message':
        continue
    input_messages.append({
        'sender': message['from'],
        'message': message['text']
    })

input_messages
```

```python
fine_tunning_data = []

start_of_text_token = "<|startoftext|>"
end_of_text_token = "<|endoftext|>"
separator_token = "<|separator|>"

for message in input_messages:
    sender = message['sender']
    message = message['message']
    input_text = f"{start_of_text_token}{sender}{separator_token}{message}{end_of_text_token}"
    fine_tunning_data.append(input_text)

fine_tunning_data
```

```python
with open("./output/finetune_text_corpus.json", 'w', encoding= "utf-8") as f:
     json.dump(fine_tunning_data, f, indent=4, ensure_ascii=False)
```

```python
text_sequences = ""
for text in fine_tunning_data:
    text_sequences += text
text_sequences
```

```python
from bpe import RegexTokenizer
tokenizer = RegexTokenizer()
tokenizer.train(text_sequences, vocab_size=1024)
```

```python
vocab = tokenizer.vocab
vocab
```

```python
tokenizer.encode("–ü—Ä–∏–≤–µ—Ç, –º–∏—Ä")
```

```python
tokenizer.decode([437, 263, 449, 314, 44, 326, 369])
```

```python
max_vocab_id = list(tokenizer.vocab.keys())[-1]
tokenizer.special_tokens = {
    "<|startoftext|>": int(max_vocab_id + 1),
    "<|separator|>": int(max_vocab_id + 2),
    "<|endoftext|>": int(max_vocab_id + 3),
    "<|unk|>": int(max_vocab_id + 4),
    "<|padding|>": int(max_vocab_id + 5),
}
```

```python
tokenizer.save('./output/tokenizer/tokenzier_v2')
```
