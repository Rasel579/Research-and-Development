```python

with open('./output/text_corpus.txt', 'r', encoding='utf-8') as f:
    text_sequences = f.read()
len(text_sequences)
```

```python
from bpe import BasicTokenizer
tokenizer = BasicTokenizer()
tokenizer.train(text_sequences, vocab_size=1024)
```

```python
vocab = tokenizer.vocab
vocab
```

```python
tokenizer.encode("Привет, мир")
```

```python
tokenizer.decode([569, 264, 430, 289, 44, 303, 497])
```

```python
max_vocab_id = list(tokenizer.vocab.keys())[-1]
tokenizer.special_tokens = {
    "<|startoftext|>": max_vocab_id + 1,
    "<|separator|>": max_vocab_id + 2,
    "<|endoftext|>": max_vocab_id + 3,
    "<|unk|>": max_vocab_id + 4,
    "<|padding|>": max_vocab_id + 5,
}
```

```python
len(tokenizer.encode(text_sequences))
```

```python
tokenizer.save('./output/tokenizer/tokenzier_v1')
```
